```python
##### Вуроятность что оба датчика отккажут #####

Pab = a * b

round(Pab, 2)
```

```python
##### Вуроятность что оба датчика сработают #####

PAB = Prob[0] * Prob[1]

PAB
```

```python
#### Вероятность что только первый сработает #######

PAb = 1 - PAB - Pab

PAb
```
