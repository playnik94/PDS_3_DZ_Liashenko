```python
import numpy as np
```

```python
M_x = 1 * 0.3 + 2 * 0.45 + 3 * 0.25
M_x
```

```python
M_y = -1 * 0.1 + 0 * 0.35 + 1 * 0.25 + 2 * 0.2 + 3 * 0.1
M_y
```

```python
D_x = 1**2 * 0.3 + 2**2 * 0.45 + 3**2 * 0.25 - M_x**2
D_x
```

```python
D_y = -1**2 * 0.1 + 0**2 * 0.35 + 1**2 * 0.25 + 2**2 * 0.2 + 3**2 * 0.1 - M_y**2
D_y
```

```python
G_x = np.sqrt(D_x)

G_y = np.sqrt(D_y)
```

```python
G_x
```

```python
G_y
```

```python

M_xy = 1*-1*0.1 + 1*0 *0.1 + 1*1*0.05 + 1*2*0.05 + 1*3*0 + 2*-1*0 + 2*0*0.2 + 2*1*0.1 + 2*2*0.1 + 2*3*0.05 + 3*-1*0 + 3*0*0.05 + 3*1*0.1 + 3*2*0.05 + 3*3*0.05
M_xy
```

```python
cov_xy = M_xy - M_x * M_y
cov_xy
```

```python
r_xy = cov_xy / (G_x * G_y)
r_xy
```

```python
M_Y = 1 * 0.25 + 2 * 0.4 + 3 * 0.35
M_Y
```

```python
D_X = -1**2 *0.4 + 1**2 *0.6 - M_X**2
D_X
```

```python
D_Y = 1**2 * 0.25 + 2**2 * 0.4 + 3**2 * 0.35 - M_Y**2
D_Y
```

```python
M_XY = -1*1*0.05 + -1*2*0.3 + -1*3*0.05 + 1*1*0.2 + 1*2*0.1 + 1*3*0.3
M_XY
```

```python
cov_XY = M_XY - M_X * M_Y
cov_XY
```

```python
r_XY = cov_XY / (D_X * D_Y)
r_XY
```
