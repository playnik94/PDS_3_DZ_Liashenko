```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import fcluster, dendrogram, linkage
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

data = ([[1, 3],
       [2, 3],
       [3, 4],
       [8, 8],
       [9, 10],
       [10, 12]])

col_names = ['Feature_1', 'Feature_2']

df = pd.DataFrame(data, columns=col_names)
```

```python
model_hierarchy = linkage(df, method='single')

dendrogram(model_hierarchy,
           orientation='top',
           distance_sort='descemding',
           show_leaf_counts=True)

plt.ylabel('distance')
plt.xlabel('point')
plt.show()
```

```python
labels = fcluster(model_hierarchy, t=df.shape[0], criterion='distance')

labels
```

```python
for i, cluster in enumerate(labels):
    print(f'data = {data[i]}, | index = {i} | cluster = {cluster} ')
```

```python
df['labels'] = labels
df
```

```python
# ANOVA для Feature1
group1_feature1 = df[df['labels'] == 1]['Feature_1']
group2_feature1 = df[df['labels'] == 2]['Feature_1']

# ANOVA для Feature2
group1_feature2 = df[df['labels'] == 1]['Feature_2']
group2_feature2 = df[df['labels'] == 2]['Feature_2']

# Применяем F-тест ANOVA
f_stat1, p_value1 = f_oneway(group1_feature1, group2_feature1)
f_stat2, p_value2 = f_oneway(group1_feature2, group2_feature2)

print(f"Feature1 - F-statistic: {f_stat1}, p-value: {p_value1}")
print(f"Feature2 - F-statistic: {f_stat2}, p-value: {p_value2}")
```

```python
# Интерпретация результата
# F-statistic: Чем больше значение, тем сильнее различие между кластерами по данному признаку.
# p-value: Показывает, насколько это различие статистически значимо. Обычно значимые различия имеют 
# 𝑝
# <
# 0.05
# p<0.05.
# В нашем примере оба признака имеют низкие 
# 𝑝
# p-значения и высокие F-статистики. Это говорит о том, что:

# Оба признака (Feature1 и Feature2) важны для разделения данных на кластеры.
# Значения сильно различаются между кластерами.
```

```python
import numpy as np

def euclidean_distance(point1, point2):
    """Вычисление Евклидова расстояния между двумя точками"""
    return np.sqrt(np.sum((point1 - point2) ** 2))

def compute_distance_matrix(data):
    """Создание матрицы расстояний для всех точек"""
    n = data.shape[0]
    dist_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(i + 1, n):
            print(f'i = {data[i]} | j = {data[j]}')
            dist_matrix[i, j] = euclidean_distance(data[i], data[j])
            dist_matrix[j, i] = dist_matrix[i, j]
    return dist_matrix

def find_closest_clusters(dist_matrix):
    """Поиск двух ближайших кластеров"""
    n = dist_matrix.shape[0]
    min_dist = np.inf
    clusters = (-1, -1)
    for i in range(n):
        for j in range(i + 1, n):
            if dist_matrix[i, j] < min_dist:
                min_dist = dist_matrix[i, j]
                clusters = (i, j)
    return clusters, min_dist

def update_distance_matrix(dist_matrix, cluster1, cluster2, linkage):
    # Определяем оставшиеся индексы после объединения кластеров
    remaining_indices = [i for i in range(len(dist_matrix)) if i not in [cluster1, cluster2]]

    # Создаем новую матрицу расстояний без строк/столбцов cluster1 и cluster2
    new_dist_matrix = dist_matrix[np.ix_(remaining_indices, remaining_indices)]

    # Вычисляем расстояния от нового кластера до остальных
    new_cluster_distances = []
    for idx in remaining_indices:
        if linkage == 'single':
            dist = np.min([dist_matrix[cluster1, idx], dist_matrix[cluster2, idx]])
        elif linkage == 'complete':
            dist = np.max([dist_matrix[cluster1, idx], dist_matrix[cluster2, idx]])
        elif linkage == 'average':
            dist = np.mean([dist_matrix[cluster1, idx], dist_matrix[cluster2, idx]])
        else:
            raise ValueError("Unsupported linkage method")
        new_cluster_distances.append(dist)

    # Добавляем новую строку/колонку для объединенного кластера
    new_cluster_distances = np.array(new_cluster_distances)
    new_dist_matrix = np.vstack([new_dist_matrix, new_cluster_distances])
    new_dist_matrix = np.hstack([new_dist_matrix, np.append(new_cluster_distances, 0).reshape(-1, 1)])

    return new_dist_matrix

def hierarchical_clustering(data, n_clusters, linkage='single'):
    """
    Реализация иерархической кластеризации
    :param data: массив данных (n_samples, n_features)
    :param n_clusters: желаемое количество кластеров
    :param linkage: правило связывания ('single', 'complete', 'average')
    :return: список кластеров
    """
    n = data.shape[0]
    clusters = [[i] for i in range(n)]
    dist_matrix = compute_distance_matrix(data)
    cluster_history = []

    while len(clusters) > n_clusters:
        (cluster1, cluster2), _ = find_closest_clusters(dist_matrix)
        cluster_history.append((clusters[cluster1], clusters[cluster2]))
        # Объединяем кластеры
        new_cluster = clusters[cluster1] + clusters[cluster2]
        clusters = [clusters[i] for i in range(len(clusters)) if i not in (cluster1, cluster2)]
        clusters.append(new_cluster)

        # Обновляем матрицу расстояний
        dist_matrix = update_distance_matrix(dist_matrix, cluster1, cluster2, linkage=linkage)

    return clusters, cluster_history
```

```python
# Пример данных
data = np.array([
    [1, 3],
    [2, 3],
    [3, 4],
    [8, 8],
    [9, 10],
    [10, 12]
])

# Иерархическая кластеризация
n_clusters = 2
clusters, history = hierarchical_clustering(data, n_clusters, linkage='average')

print("Итоговые кластеры:")
for i, cluster in enumerate(clusters):
    print(f"Кластер {i + 1}: {cluster}")

print("\nИстория объединений кластеров:")
for step, (c1, c2) in enumerate(history):
    print(f"Шаг {step + 1}: Объединяем кластеры {c1} и {c2}")
```
