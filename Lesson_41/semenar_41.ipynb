```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.info()
```

```python
df['Cabin_available'] = np.where(df['Cabin'].isnull(), 0, 1)
```

```python
df
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].replace(np.nan, 'S', inplace=True)
df['Embarked'].value_counts()
```

```python
df['Sex_num'] = np.where(df['Sex'] == 'male', 1, 0)
df.head()
```

```python
conditions = [df['Embarked'] == 'S', df['Embarked'] == 'Q', df['Embarked'] == 'C']
choices = [0, 1, 2]

df['Embarked_num'] = np.select(conditions, choices, default=0)
df.head()
```

```python
final_df = df.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)
final_df
```

```python
final_df.to_csv('final_train.csv')
```

```python
final = pd.read_csv('final_train.csv', index_col=0)
final.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
x = final.drop(['Survived'], axis=1)
y = final['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model_log = LogisticRegression()
model_log.fit(x_train, y_train)
```

```python
y_pred = model_log.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
matrix = metrics.confusion_matrix(y_test, y_pred)
matrix
```

```python
sns.heatmap(matrix, annot=True, fmt='g')
```

```python
y_pred_proba = model_log.predict_proba(x_test)[::, 1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr, tpr, label='AUC =' + str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import tree
import graphviz
```

```python
model_tree = DecisionTreeClassifier(criterion='gini', max_depth=6, min_samples_leaf=3, random_state=42)
model_tree.fit(x_train, y_train)
```

```python
predict = model_tree.predict(x_test)
print(metrics.accuracy_score(y_test, predict))
```

```python
feature_names = list(x.columns)
```

```python
feature_names
```

```python
plt.figure(figsize=(12, 8))
plot_tree(model_tree, feature_names=feature_names, class_names='Survived',
          filled=True, fontsize=10, rounded=True)
plt.show()
```

```python
target = list(str(final['Survived'].unique()))

dot_data = tree.export_graphviz(model_tree, feature_names=feature_names, class_names=target)
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models = [('LOG', LogisticRegression()), ('SGD', SGDClassifier())]
```

```python
stacking = StackingClassifier(estimators=models, final_estimator=DecisionTreeClassifier(criterion='gini', max_depth=6,                                                                                         min_samples_leaf=3, random_state=42))
stacking.fit(x_train, y_train)
```

```python
y_pred_stack = stacking.predict(x_test)
```

```python
print(metrics.accuracy_score(y_pred_stack, y_test))
print(stacking.score(x_test, y_test))
```

```python
from sklearn.ensemble import BaggingClassifier

model_bagging = BaggingClassifier(estimator=LogisticRegression(), n_estimators=10, random_state=42,)
model_bagging.fit(x_train, y_train)
```

```python
y_pred_bagging = model_bagging.predict(x_test)

print(metrics.accuracy_score(y_test, y_pred_bagging))
print(model_bagging.score(x_test, y_test))
```
