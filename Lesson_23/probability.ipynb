```python
import pandas as pd
import numpy as np

data = pd.read_csv('flights.csv', index_col=False).dropna()

### Flight startet in San Francisco ###

define_San_Francisco = (data['DestCityName'] == 'San Francisco').sum()
total_flight = len(data['DestCityName'])

print(define_San_Francisco)
print(total_flight)
print('\n')
print(f'Probability San Francisco is: {define_San_Francisco / total_flight}')

#######################################
data

data
```

```python

#### Flight started in x for all states #####

flight_states = data.groupby('DestCityName')
num_flight_per_state = flight_states.size()
flight_state_prob = num_flight_per_state.apply(lambda num_flights: num_flights / total_flight)

print(f"Flight probabiblity for each state:  {flight_state_prob}")
```

```python
print(flight_state_prob.max())
print(flight_state_prob.idxmax())
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('flights.csv', index_col=False).dropna()
data
```

```python

amster_cout = (data['OriginCityName'] == 'Amsterdam').sum()

total_city = len(data['OriginCityName'])

print(f'Amsterdam : {amster_cout}')
print()
print(f'Total count: {total_city}')
print()
print(f'Probability amsterdam is: {amster_cout / total_city}')
```

```python
flight_st = data.groupby('OriginCityName').size()
flight_st
```

```python
proba = flight_st.apply(lambda probability_city: total_city / probability_city)
print(proba)
print()
print(proba.min())
print(proba.max())
print(proba.idxmax())
print(proba.idxmin())
```

```python
#### Условная вероятность

# В урне 5 черных и 7 зеленых мячей 

# всего 12 мячей

# 1 мячем был челеный, найти вероятность того что 2 мяч тоже будет зеленым

# P(A/B) = P(AB) / P(B)

n = 12 - 1 # Так как один уже есть -зеленый
g = 7 - 1 
r = 5

round(g / n, 3)
```

```python
#### Теорема умножения

# Кол во вопрсов 75

# Кол во ответов 40

# найти вероятность что ответит на 3 вопроса

# P(AB) =  P(A) * P(B)

a = 40 / 75 * 39 / 74 * 38 / 73
round(a, 3)

########################
```

```python
#### Теорема сложения

## Найти вероятность того что при подкидывании 2 игралтных кубиков хотя бы один раз выпадин 6 очков

# P(A + B) = P(A) + P(B) - P(AB)

# P(A) = 1/6

# P(B) = 1/6

p = 1/6 + 1/6 - 1/36
round(p, 3)
```

```python
### Вероятность противоположного события

# P(A) + P(A!) = 1

# P(A!) = 1 - P(A)
```

```python
# D = A1 - 0.8,  A2 - 0.9,  A3 - 0.85

# D = A!1 * A!2 * A!3 = 1 - 0.8 * 1 - 0.9 * 1 - 0.85 

D = 0.2 * 0.1 * 0.15

round(D, 3) # Договор не выполгнит ни одно придприятие
```
