```python
import numpy as np

class XGBoostRegressor:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.models = []

    def fit(self, X, y):
        # Инициализируем предсказания как среднее значение y
        y_pred = np.full(y.shape, y.mean())
        
        for i in range(self.n_estimators):
            # Вычисляем остатки
            residuals = y - y_pred
            
            # Создаем новое дерево и обучаем его на остатках
            tree = self.build_tree(X, residuals, depth=0)
            self.models.append(tree)
            
            # Обновляем предсказания, добавляя результат текущего дерева
            y_pred += self.learning_rate * self.predict_tree(tree, X)

    def build_tree(self, X, y, depth):
        # Простая реализация бинарного дерева
        if depth >= self.max_depth or len(set(y)) == 1:
            return np.mean(y)

        # Выбираем лучший признак и порог для разделения данных
        best_feature, best_threshold, min_loss = None, None, float("inf")
        for feature in range(X.shape[1]):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                left_mask = X[:, feature] <= threshold
                right_mask = X[:, feature] > threshold
                if sum(left_mask) == 0 or sum(right_mask) == 0:
                    continue

                left_y, right_y = y[left_mask], y[right_mask]
                loss = self.calculate_loss(left_y, right_y)

                if loss < min_loss:
                    min_loss = loss
                    best_feature = feature
                    best_threshold = threshold

        # Разделяем данные и строим поддеревья
        left_mask = X[:, best_feature] <= best_threshold
        right_mask = X[:, best_feature] > best_threshold
        left_tree = self.build_tree(X[left_mask], y[left_mask], depth + 1)
        right_tree = self.build_tree(X[right_mask], y[right_mask], depth + 1)

        return (best_feature, best_threshold, left_tree, right_tree)

    def calculate_loss(self, left_y, right_y):
        # Простая квадратичная ошибка для каждой части
        loss = np.mean((left_y - left_y.mean()) ** 2) + np.mean((right_y - right_y.mean()) ** 2)
        return loss

    def predict_tree(self, tree, X):
        # Рекурсивно вычисляем предсказание для каждого объекта
        feature, threshold, left, right = tree
        if isinstance(left, float) and isinstance(right, float):
            return np.where(X[:, feature] <= threshold, left, right)
        predictions = np.empty(X.shape[0])
        left_mask = X[:, feature] <= threshold
        right_mask = X[:, feature] > threshold
        predictions[left_mask] = self.predict_tree(left, X[left_mask])
        predictions[right_mask] = self.predict_tree(right, X[right_mask])
        return predictions

    def predict(self, X):
        # Суммируем предсказания всех деревьев с учетом learning_rate
        y_pred = np.zeros(X.shape[0])
        for tree in self.models:
            y_pred += self.learning_rate * self.predict_tree(tree, X)
        return y_pred
```

```python
import numpy as np

class XGBoostClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.models = []

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def fit(self, X, y):
        # Инициализируем предсказания как логарифм шансов для каждого класса
        y_pred = np.full(y.shape, np.log(np.mean(y) / (1 - np.mean(y))))
        
        for i in range(self.n_estimators):
            # Вычисляем псевдо-остатки (градиенты)
            residuals = y - self.sigmoid(y_pred)
            
            # Создаем новое дерево и обучаем его на градиентах
            tree = self.build_tree(X, residuals, depth=0)
            self.models.append(tree)
            
            # Обновляем предсказания
            y_pred += self.learning_rate * self.predict_tree(tree, X)

    def build_tree(self, X, y, depth):
        # Простая реализация бинарного дерева
        if depth >= self.max_depth or len(set(y)) == 1:
            return np.mean(y)

        # Выбираем лучший признак и порог для разделения данных
        best_feature, best_threshold, min_loss = None, None, float("inf")
        for feature in range(X.shape[1]):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                left_mask = X[:, feature] <= threshold
                right_mask = X[:, feature] > threshold
                if sum(left_mask) == 0 or sum(right_mask) == 0:
                    continue

                left_y, right_y = y[left_mask], y[right_mask]
                loss = self.calculate_loss(left_y, right_y)

                if loss < min_loss:
                    min_loss = loss
                    best_feature = feature
                    best_threshold = threshold

        # Разделяем данные и строим поддеревья
        left_mask = X[:, best_feature] <= best_threshold
        right_mask = X[:, best_feature] > best_threshold
        left_tree = self.build_tree(X[left_mask], y[left_mask], depth + 1)
        right_tree = self.build_tree(X[right_mask], y[right_mask], depth + 1)

        return (best_feature, best_threshold, left_tree, right_tree)

    def calculate_loss(self, left_y, right_y):
        # Простая квадратичная ошибка для каждой части
        loss = np.mean((left_y - left_y.mean()) ** 2) + np.mean((right_y - right_y.mean()) ** 2)
        return loss

    def predict_tree(self, tree, X):
        # Рекурсивно вычисляем предсказание для каждого объекта
        feature, threshold, left, right = tree
        if isinstance(left, float) and isinstance(right, float):
            return np.where(X[:, feature] <= threshold, left, right)
        predictions = np.empty(X.shape[0])
        left_mask = X[:, feature] <= threshold
        right_mask = X[:, feature] > threshold
        predictions[left_mask] = self.predict_tree(left, X[left_mask])
        predictions[right_mask] = self.predict_tree(right, X[right_mask])
        return predictions

    def predict_proba(self, X):
        # Суммируем предсказания всех деревьев и применяем сигмоиду
        y_pred = np.zeros(X.shape[0])
        for tree in self.models:
            y_pred += self.learning_rate * self.predict_tree(tree, X)
        return self.sigmoid(y_pred)

    def predict(self, X):
        # Преобразуем вероятности в классы
        return (self.predict_proba(X) > 0.5).astype(int)
```

```python
import numpy as np
from sklearn import metrics
from sklearn.datasets import make_regression, make_classification
from xgboost import XGBClassifier, XGBRegressor
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x, y = make_regression(n_samples=500, n_features=1, n_informative=1, noise=55)
X, Y = make_classification(n_samples=500, n_features=1, n_informative=1,
                           n_redundant=0, n_repeated=0, n_clusters_per_class=1, weights=[0.2, 0.8])
```
