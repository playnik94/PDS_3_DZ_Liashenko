```python
import numpy as np 
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from numpy import random
from sklearn.mixture import GaussianMixture
import pandas as pd


x, y = make_blobs(n_samples=(8000, 5000), cluster_std=(0.5, 0.7), center_box=(9.8, 15))
```

```python
x
```

```python
model_Ga = GaussianMixture(n_components=2).fit(x)
```

```python
center = model_Ga.means_
center
```

```python
predicts = model_Ga.predict(x)
```

```python
predicts
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], color='lightblue', label='x')
plt.scatter(center[:,0], center[:,1], color='blue', label='center')
plt.legend()
plt.show()
```

```python
from pandas import DataFrame

data = DataFrame({'X':x[:,0], 'Y':x[:,1], 'label':predicts})
data_g = data.groupby('label')
```

```python
fig, ax = plt.subplots()
for name, group in data_g:
    ax.scatter(group.X, group.Y, label=name)
    ax.scatter(center[:,0], center[:,1])

ax.legend()
plt.show()
```
